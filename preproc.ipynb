# Merge files into one

```python
# mirai has indecies 1-115
# others have 0-114

column_names = ['feature_'+str(i) for i in range(0,116)]
column_names[-1] = 'x'

df1.columns = column_names
df2.columns = column_names
df3.columns = column_names
df4.columns = column_names
df5.columns = column_names
df6.columns = column_names
df7.columns = column_names
df8.columns = column_names
df9.columns = column_names

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], axis=0)
print(f'dataset shape: {df.shape}')
df.head()
```

# Preprocessing

```python
df.describe()
```

### Handle missing data

```python
#check null values
df_null = df[df.isna().any(axis=1)]
df_null
```

### Handle outliers

```python
from scipy import stats

z_scores = np.abs(stats.zscore(df, nan_policy='omit'))
z_scores.head()
```

```python
outliers_threshold = 3
mask = (z_scores <= outliers_threshold).all(axis=1)
print(f" No. of outlier rows = {len(mask[mask == False])} from (total) {df.shape[0]} rows")
```

```python
df_zscored = df[mask]
print(f'dataset shape after droping outliers: {df_zscored.shape}')
df_zscored.head()
```

```python
# Save
df_zscored.to_csv('processed_dataset.csv', index=False)
```

### Standardization
Because tree-based classification algorithms will be used, it is not needed to perform feature scaling.
However here is the example (in case we will need one later)

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

df_scaled = MinMaxScaler().fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns = list(df))
df_scaled.describe()
```

```python
df_std = StandardScaler().fit_transform(df_scaled)
df_std = pd.DataFrame(df_std, columns = list(df_scaled))
df_std.describe()
```

```python
z_scores = np.abs(stats.zscore(df_std, nan_policy='omit'))
z_scores.head()
```

```python
outliers_threshold = 3
mask = (z_scores <= outliers_threshold).all(axis=1)
print(f" No. of outlier rows = {len(mask[mask == False])} from (total) {df_std.shape[0]} rows")
```

```python
df_zscored = df_std[mask]
print(f'dataset shape after droping outliers: {df_zscored.shape}')
df_zscored.head()
```

```python
# Save
df_zscored.to_csv('processed_scaled_dataset.csv', index=False)
```
